### Предобработка данных

В данных много пропусков, пройдемся по столбцам, чтобы понять, как заменить пропуски

#### ceiling_height

```python
data.hist('ceiling_height', range=(0,8), bins=20, figsize=(4,4))
```

Как видно из гистограммы, наиболее распространено стандартное значение высоты потолка между 2 и 3 метрами. Заменим пропуски в высоте потолка на медианное значение

```python
data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
```

#### floors_total

Проверим, какие значения этажей встречаются в тех строках, где не указано общее число этажей в доме. Нельзя просто заменить все пустые значения медианой (9 этажей), потому что в таком случае число этажей в доме может стать меньше эатажа, на котором расположена квартира. Но нужно удостовериться, что значения этажей в строках с пустым числом этажей в доме в принципе адекватны

```python
data.loc[data['floors_total'].isna(), 'floor'].unique()
```

Значения этажей адекватны, значит можно придумать алгоритм замены пустых значений. Предположим, что в домах число этажей кратно 5 (5, 10, 15 и т.д.)

```python
# функция для округления в высшую сторону до ближайшего числа, кратного base
def round_up_to_base(number, base):
    return base * (int(number / base) + (number % base > 0))
```

```python
data.loc[(data['floors_total'].isna()) & (data['floor'] <= data['floors_total'].median()), 'floors_total'] = \
data['floors_total'].median()

data.loc[data['floors_total'].isna(), 'floors_total'] = \
data.loc[data['floors_total'].isna(), 'floor'].apply(lambda x: round_up_to_base(x, 5))
```

```python
data.query('floors_total < floor').shape[0]
```

#### living_area

Сгруппируем квартиры по числу комнат и в каждой группе рассчитаем соотношение медианы жилой площади и медианы общей площади. Умножим получившийся показатель на коэффициент 1.25 (корректировка на тот случай, чтобы не завышалась жилая площадь). На медиану по группе заменить не так хорошо, потому что есть очень большие квартиры, а мединная кухня будет слишком мала для них

```python
for rooms in data['rooms'].unique():
    median_living_area = data.loc[(data['rooms'] == rooms), 'living_area'].median()
    median_total_area = data.loc[(data['rooms'] == rooms), 'total_area'].median()
    total_to_living = 1.25 * median_total_area / median_living_area
    
    data.loc[(data['rooms'] == rooms) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['rooms'] == rooms) & (data['living_area'].isna()), 'total_area'] / total_to_living
```

#### is_apartment

```python
data['is_apartment'].value_counts()
```

Аппартаментов в принципе мало, это подтверждается тем, что значений False намного больше True. И скорее всего, в тех строках, где встречается пустота, тоже должно стоять False

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
data['is_apartment'].value_counts()
```

#### kitchen_area

Сгруппируем квартиры по числу комнат и в каждой группе рассчитаем соотношение площади кухни и общей площади и рассчитаем исходя из этого коэффициента новые значения.

В данных есть строки, в которых число комнат равно 0. Полагаю, что 0 в таком случае можно мысленно принять за 1, поскольку даже если речь о квартирах со свободной планировкой, изначально в них все равно 1 большая комната.

```python
data.loc[(data['rooms'] == 0), 'rooms'] = 1
```

```python
for rooms in data['rooms'].unique():
    median_total_area = data.loc[(data['rooms'] == rooms), 'total_area'].median()
    median_kitchen_area = data.loc[(data['rooms'] == rooms), 'kitchen_area'].median()
    kitchen_coeff = 1.25 * median_total_area / median_kitchen_area
    
    data.loc[(data['rooms'] == rooms) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['rooms'] == rooms) & (data['kitchen_area'].isna()), 'total_area'] / kitchen_coeff
```

```python
data.loc[data['kitchen_area'].isna()].shape[0]
```

#### ближайшие объекты

Разобьем строки на группы в зависимости от населенного пункта. Можно попробовать заменить все пустые значения на медианные значения аналогичных показателей в группах, потому что в принципе квартиры в рамках одного населенного пункта (особенно в рамках сел и деревень) в среднем практически одинаково равноудалены от ближайших озер, парков и т.д.

```python
data.loc[data['cityCenters_nearest'].isna(), 'locality_name'].value_counts()
```

```python
for locality in data['locality_name'].unique():
    data.loc[(data['locality_name'] == locality) & (data['cityCenters_nearest'].isna()), 'cityCenters_nearest'] = \
    data.loc[(data['locality_name'] == locality), 'cityCenters_nearest'].median()
    
    data.loc[(data['locality_name'] == locality) & (data['airports_nearest'].isna()), 'airports_nearest'] = \
    data.loc[(data['locality_name'] == locality), 'airports_nearest'].median()
    
    data.loc[(data['locality_name'] == locality) & (data['parks_nearest'].isna()), 'parks_nearest'] = \
    data.loc[(data['locality_name'] == locality), 'parks_nearest'].median()
    
    data.loc[(data['locality_name'] == locality) & (data['ponds_around3000'].isna()), 'ponds_around3000'] = \
    data.loc[(data['locality_name'] == locality), 'ponds_around3000'].median()
    
    data.loc[(data['locality_name'] == locality) & (data['ponds_nearest'].isna()), 'ponds_nearest'] = \
    data.loc[(data['locality_name'] == locality), 'ponds_nearest'].median()
    
    data.loc[(data['locality_name'] == locality) & (data['parks_around3000'].isna()), 'parks_around3000'] = \
    data.loc[(data['locality_name'] == locality), 'parks_around3000'].median()
```
