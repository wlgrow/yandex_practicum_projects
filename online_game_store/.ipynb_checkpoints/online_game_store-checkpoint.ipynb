Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Выберем одну из популярных платформ

```python
PS4_data = data_actual.query('platform == "PS4"')
```

Построим диаграмму рассеяния по оценкам критиков и суммарным продажам

Коэффициент корреляции показывает наличие линейной связи между признаками. Согласно графикам и коэффициенту корреляции существует средняя (скорее слабая) прямая связь между оценками критиков и продажами игры. Но в то же время прямая связь между оценками пользователей и продажами не прослеживается (коэффициент близок к нулю). Получается, пользователи в своих оценках менее разборчивы, чем критики.

Однако из формы графиков можно сделать вывод, что нелинейная связь между показателями, скорее всего, существует (возможно, закон описывает какая-либо степенная функция).

Посчитаем аналогичные коэффициенты корреляции для других платформ.

```python
for platform in data_actual['platform'].unique()[:7]:
    platform_data = data_actual.query('platform == @platform')
    platform_data.plot(x='critic_score', y='all_sales', kind='scatter', figsize=(3,3), title=platform)
    print(f'Корр. критики и продажи ({platform}): ', 
          platform_data['critic_score'].corr(platform_data['all_sales']))
    platform_data.plot(x='user_score', y='all_sales', kind='scatter', figsize=(3,3),title=platform)
    print(f'Корр. пользователи и продажи ({platform}): ', 
          platform_data['user_score'].corr(platform_data['all_sales']))
    print('-'*20)
```

По ряду других платформ наблюдается схожая ситуация: корреляция оценок критиков с продажами сильнее, чем корреляция пользовательских оценок с продажами (на диаграммах рассеяния оценки пользователей более хаотичны, чем оценки критиков). Выделяется только платформа WiiU - корреляция оценок игроков с продажами тут выше, чем корреляция оценок критиков с продажами.

### Портрет пользователя каждого региона

#### Самые популярные платформы

Напишем функцию, которая будет отбрасывать все платформы не в топ5 в категорию other

```python
def platform_top(platform, top5):
    if platform in top5:
        return platform
    else:
        return "other"
```

Найдем топ-5 платформ для каждого региона

```python
top5dict = dict()
```

```python
for region in ['na_sales', 'eu_sales', 'jp_sales']:
    top5 = (data_actual
            .pivot_table(index='platform', values=[region], aggfunc='sum')
            .sort_values(by=region, ascending = False)
            .head(5)
            .index.values
           )
    top5dict[region] = top5  
```

```python
top5dict
```

Построим пайчарты для каждого региона с учетом топ-5 платформ для этого региона

```python
fig, ax = plt.subplots(nrows=1, ncols=3)

fig.set_size_inches(12, 4)
fig.subplots_adjust(wspace=0.2)

index = 0

for region in ['na_sales', 'eu_sales', 'jp_sales']:
    data_reg = data_actual.copy()
    data_reg['platform_top'] = data_reg['platform'].apply(platform_top, args=(top5dict[region],))
    (data_reg
     .pivot_table(index='platform_top', values=[region], aggfunc='sum')
     .plot.pie(y=region, figsize=(12, 6), ax=ax[index], legend=False, autopct="%.f%%")
    )
    index += 1
```

Если Топ-5 консолей в Северной Америке и Европе примерно одинаковы, то в Японии очень велика доля портативных консолей. Это связано с образом жизни японцев: например, в крупных городах они много времени проводят в транспорте, часто вынуждены играть на ходу.

```python
def genre_top(genre, top5):
    if genre in top5:
        return genre
    else:
        return "other"
```

```python
top5dict = dict()
for region in ['na_sales', 'eu_sales', 'jp_sales']:
    top5 = (data_actual
            .pivot_table(index='genre', values=[region], aggfunc='sum')
            .sort_values(by=region, ascending = False)
            .head(5)
            .index.values
           )
    top5dict[region] = top5
    
top5dict
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3)

fig.set_size_inches(12, 4)
fig.subplots_adjust(wspace=0.2)

index = 0

for region in ['na_sales', 'eu_sales', 'jp_sales']:
    data_reg = data_actual.copy()
    data_reg['genre_top'] = data_reg['genre'].apply(platform_top, args=(top5dict[region],))
    (data_reg
     .pivot_table(index='genre_top', values=[region], aggfunc='sum')
     .plot.pie(y=region, figsize=(12, 6), ax=ax[index], legend=False, autopct="%.f%%")
    )
    index += 1
```

В Северной Америке и Европе на первых местах такие жанры, как Action, Shooter, Sports - игры этих жанров распространены на консолях, популярных в указанных регионах. В Японии шутеры вообще не входят в топ-5 жанров. А первое место с большим отрывом занимают ролевые игры (все логично, это, действительно, очень популярный игровой жанр. ### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем гипотезы:
H0 -Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H1 - Средние пользовательские рейтинги платформ Xbox One и PC не одинаковые

```python
xbox = data_actual.loc[(data_actual['platform'] == 'XOne') & (data_actual['user_score'].notna()), 'user_score']
print(xbox.shape)

pc = data_actual.loc[(data_actual['platform'] == 'PC') & (data_actual['user_score'].notna()), 'user_score']
print(pc.shape)
```

Гипотеза H0 двухсторонняя, поскольку средние рейтинги пользователей двух выбранных платформ могут отличаться в разные стороны (одна больше или меньше другой)

Применим t-тест, исходя из того, что:
- Генеральные совокупности не зависят друг от друга.
- Выборки больше 30: возьмем одинаковые выборки по 200 строк - плюс у нас нет оснований полагать, что дисперсии ген совокупностей не равны

```python
alpha = .05

results = st.ttest_ind(
    xbox,
    pc)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
xbox.mean()
```

```python
pc.mean()
```

Если предположить, что средние двух совокупностей равны, вероятность получить такое или большее различие составила около 55%. 
Мы не можем отвергнуть гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC.

В реальности средние значения по платформам, действительно, близки.

#### Средние пользовательские рейтинги жанров Action и Sports разные

Нулевая гипотеза всегда формулируется через равенство, поэтому примем за H0 гипотезу "Средние пользовательские рейтинги жанров Action и Sports равны". H1 в этом случае будет "Средние пользовательские рейтинги жанров Action и Sports различаются"

Снова имеем двустороннюю гипотезу

```python
action = data_actual.loc[(data_actual['genre'] == 'Action') & (data_actual['user_score'].notna()), 'user_score']
print(action.shape)

sports = data_actual.loc[(data_actual['genre'] == 'Sports') & (data_actual['user_score'].notna()), 'user_score']
print(sports.shape)
```

Применим t-тест, исходя из того, что:
- Генеральные совокупности не зависят друг от друга.
- Выборки больше 30: возьмем одинаковые выборки по 700 строк - плюс у нас нет оснований полагать, что дисперсии ген совокупностей не равны

```python
alpha = .05

results = st.ttest_ind(
    action,
    sports)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Если предположить, что средние двух совокупностей равны, вероятность получить такое или большее различие составила очень маленькое число, близкое к 0. 
Мы должны отвергнуть гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports

В реальности средние значения по платформам, действительно, различаются.

```python
action.mean()
```

```python
sports.mean()
```
